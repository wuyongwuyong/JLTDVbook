# **快速开始**

本地不再安装的使用方法，只需要下载好VScode能登陆服务器即可。

## **VS code下载**

下载好VS code之后进行相关基本配置，可以参考[该网址](https://blog.csdn.net/zhaxun/article/details/120568402)进行配置。

```{admonition} 提示
:class: tip
注意安装好SSH远程插件，以及确保是上海交通大学的校园网，如在校外请使用[交大VPN](https://net.sjtu.edu.cn/wlfw/VPN.htm)。
```

### **IP配置**

**VS code**上连接服务器的**IP**配置如下：

```cfg
Host Jmatlab
    HostName 202.120.39.189
    User Jmatlab
    Port 60003
```

```{admonition} 提示
:class: tip
直接在原有的`SSH config`文件下面增加上面的语句后刷新即可，原来的一些远程连接都能在列表观察到。Host自己想改啥改啥。
```

## **JUPYTER BOOK编译**

由于markdown有不同的语法比如`MyST`、`MyST-nb`等，在**VS code**中直接看ipynb文件有些结构无法预览，比如[下面章节的框架](content:ref:label1)。因此最好的预览方式是更改完`.ipynb`文件后，编译成`.html`文件进行预览。

一些配置都已经配置好，需要做的是找到**LTDVbook**的路径，然后在**VS code**中打开该路径下的**ipynb**文件，然后按照下面的步骤进行编译。

```{admonition} 备注

整个Laser Team Data Viewer的**Jupyter book**项目路径为`/home/Jmatlab/workspace/LTDV/LTDVbook/LTDVbook`。
```

添加好`_toc.yml`文件后，使用[Live Server](Content:LiveServer:ref)打开`~/_build/html/index.html`文件，直接运行下列语句即可新编译，刷新`.html`就能直接看到更改。

## **MATLAB代码框**


### **MATLAB基本代码框**

在**VS code**中添加代码框，注意选择Kernel为**MATLAB**核，在代码框中写入**MATLAB**代码即可，如下所示。

`````matlab
%plot -s 400,300   
%上述表示限制输出的图像大小尺寸

clear,clc
x = linspace(0,20,1000); 
y1 = 2*cos(x);
y2 = sin(2*x);
y3 = 3*sin(x);
f1 = figure(1); 
plot(x,y1);
hold on;
plot(x,y2);
plot(x,y3);

grid on;
`````

(content:ref:func)=
### **Cell**里面书写**MATLAB**函数

此外，一些notebook中的MATLAB操作可以参考[该文档](https://am111.readthedocs.io/en/latest/jmatlab_use.html#Plotting)。下列示例表示如何在Cell中使用自定义的Function函数。


`````matlab
%%file ./LTDV/MATLABcode/Wuyong/multi_line_func.m
%上述表示创建一个.m文件，注意文件名和函数名一致

function p = multi_line_func(a,b)
    % in-file comments can be added like this
    a = a+1;
    b = b+1;
    p = a+b;
end
`````

`````matlab
multi_line_func(1,1)
`````
